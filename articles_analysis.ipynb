```python
import re

def extract_links_from_line(line):
    # Regular expression pattern to match URLs
    url_pattern = r'(https?://\S+)'
    # Extract all URLs from the line using the pattern
    urls = re.findall(url_pattern, line)
    urls = [i for i in urls]
    return urls

def extract_links_from_file(file_path):
    links = []
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()  # Remove leading/trailing whitespace and newline characters
            urls = extract_links_from_line(line)
            links.extend(urls)
    return links

# all astro links
links = extract_links_from_file(file_path)
links
```

```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_website(url):
    # Send a GET request to the website
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Create a BeautifulSoup object to parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract text from the parsed HTML
        text = soup.get_text(separator=' ')

        return text

    # If the request was not successful, return None or handle the error as needed
    return None
```

```python
from waybackpy import WaybackMachineCDXServerAPI

def get_wayback_url(url):
  user_agent = "Mozilla/5.0 (Windows NT 5.1; rv:40.0) Gecko/20100101 Firefox/40.0"
  cdx = WaybackMachineCDXServerAPI(url, user_agent, start_timestamp=2020, end_timestamp=2023)
  return [str(i) for i in cdx.snapshots()][:-1]
```
